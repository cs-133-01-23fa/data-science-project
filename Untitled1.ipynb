```python
from sklearn import datasets
import pandas as pd

# Load iris and digits datasets
iris = datasets.load_iris()
digits = datasets.load_digits()
diabetes = datasets.load_diabetes()
# Replace 'path_to_csv' with the actual path to the Boston dataset CSV
boston = pd.read_csv('/content/boston_house_prices.csv')
health = pd.read_csv('/content/Mental Health Care in the Last 4 Weeks.csv')
# Accessing the data
iris_data = iris.data
digits_data = digits.data
boston_data = boston.values
diabetes_data = diabetes.data
health_data = health.values
# Print shapes
print("Iris data shape:", iris_data.shape)
print("Digits data shape:", digits_data.shape)
print("Boston data shape:", boston_data.shape)
print("Diabetes data shape:",diabetes_data.shape)
print("Health data shape:",health_data.shape)
# Print Size
print("Iris data size:", iris_data.size)
print("Digits data size:", digits_data.size)
print("Boston data size:", boston_data.size)
print("Diabetes data size:",diabetes_data.size)
print("Health data size:", health_data.size)
#Print dataset
print("Iris data:", iris_data)
print("Digits data:", digits_data)
print("Boston data:", boston_data)
print("Diabetes data:",diabetes_data)
print("Health data:", health_data)
```
